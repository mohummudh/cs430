# Question 1
The linear relationship between diameter and length.

```python
X, y = df['Length'].values.reshape(-1, 1), df['Diameter'].values.reshape(-1, 1)
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
pred = model.predict(X)
correlation = np.corrcoef(pred.flatten(), y.flatten())
```

```python
plt.scatter(X, y, alpha =0.5, label='Data')
plt.plot(X, pred, color='red', label='Linear Fit')
plt.xlabel('Length')
plt.ylabel('Diameter')
plt.legend()
plt.title("Linear relationship between abalone length and diameter")
plt.show()
print(f"The intercept of the line is {model.intercept_[0]}, with a gradient of {model.coef_[0][0]}.")
print(f"The correlation coefficient is {correlation[0,1]}.")
```

# Question 2

```python
mulModel.fit(Xmul,ymul)
```

```python
plt.scatter(0.93656021 * X1.flatten() + 1.11164951 * X2.flatten() + 1.25296164 * X3.flatten(), ymul)
plt.ylabel("Whole Weight")
plt.xlabel("Sum of weights")
print("The relationship holds, but is not always exact, there may be other factors adding to the weight.")
print(f"The intercept is {mulModel.intercept_[0]}, with parameters for shucked weight, viscera weight and shell weight: {mulModel.coef_[0]}")
```

# Question 3

## Question 3a
The linear model!

```python
model1 = LinearRegression()
```

```python
model1.fit(diameter, weight)
```

```python
pred1 = model1.predict(diameter)
```

```python
correlation1 = np.corrcoef(weight.flatten(), pred1.flatten())
r1 = r2_score(weight.flatten(), pred1.flatten())
```

```python
plt.scatter(diameter, weight, alpha =0.5, label='Data')
plt.plot(diameter.flatten(), pred1.flatten(), color='red', label='Linear Fit')
plt.xlabel('Diameter')
plt.ylabel('Whole Weight')
plt.legend()
plt.show()

print(f"The intercept is {model1.intercept_[0]}, with a gradient of {model1.coef_[0][0]}.")
print(f"The correlation coefficient is {correlation1[0,1]}.")
print(f"The R^2 is {r1}.")
```

## Question 3b
The quadratic model.

```python
poly = PolynomialFeatures(degree=2)
qdiameter = poly.fit_transform(diameter)
```

```python
model2 = LinearRegression()
model2.fit(qdiameter, weight)
```

```python
pred2 = model2.predict(qdiameter)
```

```python
correlation2 = np.corrcoef(weight.flatten(), pred2.flatten())
r2 = r2_score(weight.flatten(), pred2.flatten())
```

```python
plt.scatter(diameter, weight, alpha =0.5, label='Data')
sns.lineplot(x=diameter.flatten(), y=pred2.flatten(), color='red', label='Quadratic Model')
# plt.scatter(diameter.flatten(), pred2.flatten(), color='red', label='Quadratic Fit')
plt.xlabel('Diameter')
plt.ylabel('Whole Weight')
plt.legend()
plt.show()

print(f"The intercept is {model2.intercept_[0]}, with a coefficients of {model2.coef_}")
print(f"The correlation coefficient is {correlation2[0,1]}.")
print(f"The R^2 is {r2}.")
```

## Question 3c
The cubic model!

```python
cube = diameter**3
model3 = LinearRegression()
model3.fit(cube, weight)
model3.intercept_ = 0
```

```python
pred3 = model3.predict(cube)
```

```python
correlation3 = np.corrcoef(weight.flatten(), pred3.flatten())
r3 = r2_score(weight.flatten(), pred3.flatten())
```

```python
plt.scatter(diameter, weight, alpha =0.5, label='Data')
sns.lineplot(x=diameter.flatten(), y=pred3.flatten(), color='red', label='Cubic model')

plt.xlabel('Diameter')
plt.ylabel('Whole Weight')
plt.legend()
plt.show()

print(f"A coefficient of {model3.coef_}")
print(f"The correlation coefficient is {correlation3[0,1]}.")
print(f"The R^2 is {r3}.")
```

```python
model4 = LinearRegression()
model4.fit(diameter, logw)
```

```python
# plt.scatter(diameter, logw, alpha =0.5, label='Data')
# sns.lineplot(x=diameter.flatten(), y=pred4.flatten(), color='red', label='Line Plot')
# # plt.scatter(diameter.flatten(), pred2.flatten(), color='red', label='Quadratic Fit')
# plt.xlabel('Diameter')
# plt.ylabel('ln( Whole Weight )')
# plt.legend()
# plt.show()
# print(f"The intercept is {model4.intercept_[0]}, with coefficients of {model4.coef_}")
```

```python
# correlation4 = np.corrcoef(logw.flatten(), pred4.flatten())
# r4 = r2_score(weight.flatten(), pred3.flatten())
# print(f"The correlation coefficient is {correlation4[0,1]}.")
# print(f"The R^2 is {r4}.")
```

```python
alpha = np.exp(-3.7509786563887033)
l = 8.11667132
predictions = alpha * np.exp(l*diameter)
```

```python
correlation4 = np.corrcoef(weight.flatten(), predictions.flatten())
r4 = r2_score(weight.flatten(), predictions.flatten())
```

```python
plt.scatter(diameter, weight, alpha =0.5, label='Data')
sns.lineplot(x=diameter.flatten(), y=predictions.flatten(), color='red', label='Exponential Model')
# plt.scatter(diameter.flatten(), pred2.flatten(), color='red', label='Quadratic Fit')
plt.xlabel('Diameter')
plt.ylabel('Whole Weight')
plt.legend()
plt.show()
print(f"The model is y = {alpha} * exp({l} * x)")
print(f"The correlation coefficient is {correlation4[0,1]}.")
print(f"The R^2 is {r4}.")
```

Comparing visually and using the calculated coefficients, the cubic model seems the most accurate. This may be because the weight may be related to the volume of the abalone, which is related to the diameter.
