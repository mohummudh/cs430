## Question 1

```python
diameter = df["Diameter"]
length = df["Length"]
```

```python
type(diameter.values)
```

```python
plt.scatter(length.values, diameter.values)
plt.show()
```

```python
# linear regression fit
# note that the x vector needs to be reshaped as a column
model = LinearRegression()

model.fit(length.values.reshape((-1, 1)), diameter.values)
```

```python
# use the model to predict the y values of the inputs
y_pred = model.predict(length.values.reshape((-1, 1)))
```

```python
plt.scatter(length.values, diameter.values)
plt.plot(length.values, y_pred, color='red')
plt.xlabel("Length")
plt.ylabel("Diameter")
plt.title("Question 1")
plt.show()
```

### Results from the linear fit

```python
# get the parameters of the model
params = model.get_params()

# compute the R^2 score of the model
r2 = model.score(length.values.reshape((-1, 1)), diameter.values)
corr = np.corrcoef(length.values, diameter.values)

print(f"The gradient of the linear fit is {model.coef_} and the y_intercept is {model.intercept_}")
print(f"R^2 values is {r2}")
print(f"Correlation coefficient is {corr}")
```

The value for the gradient suggests a strong positive correlation between the length of Abalone and its diameter, i.e. the diameter of Abalone increases with its length.

```python
# feature engineering
df["Infant/Adult"] = df["Sex"].apply(lambda x: 0 if x == "I" else 1)
df
```

Our new column has value $0$ if the abalone is an infant and has the value $1$ if the abalone is an adult.

```python
from sklearn.linear_model import LogisticRegression

# part (a) length only
clf1 = LogisticRegression().fit(df["Length"].values.reshape(-1,1), df["Infant/Adult"])

# part (b) whole weight only
clf2 = LogisticRegression().fit(df["Whole weight"].values.reshape(-1,1), df["Infant/Adult"])

# part (c) class rings only
clf3 = LogisticRegression().fit(df["Rings"].values.reshape(-1,1), df["Infant/Adult"])

# part (d) all of the above together
clf4 = LogisticRegression().fit(df[["Length", "Whole weight", "Rings"]].values, df["Infant/Adult"])
```

```python
# accuracy values
acc1 = clf1.score(df["Length"].values.reshape(-1,1), df["Infant/Adult"])
acc2 = clf2.score(df["Whole weight"].values.reshape(-1,1), df["Infant/Adult"])
acc3 = clf3.score(df["Rings"].values.reshape(-1,1), df["Infant/Adult"])
acc4 = clf4.score(df[["Length", "Whole weight", "Rings"]].values, df["Infant/Adult"])

print(f"The accuracy of model 1 is {acc1}.")
print(f"The accuracy of model 2 is {acc2}.")
print(f"The accuracy of model 3 is {acc3}.")
print(f"The accuracy of model 4 is {acc4}.")
```

From the above accuracy scores, it appears that the best logistic regressor is the one that takes multiple relevant features into account at once.

```python
def normalised_bar_charts(df, index, title):
    pivot_table = pd.pivot_table(df, index=index, columns=df["gender"], aggfunc='size', fill_value=0)
    pivot_table["Female"] = pivot_table["Female"] / pivot_table["Female"].sum()
    pivot_table["Male"] = pivot_table["Male"] / pivot_table["Male"].sum()
    
    # plot bar chart
    pivot_table.plot(kind='bar')
    plt.xlabel('Category')
    plt.xticks(rotation=60)
    plt.ylabel('Proportion')
    plt.title(title)
    plt.legend(title='Group')

    plt.show()
```

```python
features = ["marital-status", "occupation", "relationship", "hours-per-week", "capital-gain", "capital-loss", "income"]
```

```python
# label encoding categorical features

from sklearn.preprocessing import LabelEncoder

categorical_features = ["workclass", "education", "marital-status", "occupation", "relationship", "race", "native-country", "income"]

label_encoder = LabelEncoder()

# make a copy of the dataframe
df_enc = df.copy()

for feature in categorical_features:
    # Assuming 'df' is your DataFrame containing the categorical feature
    df_enc[feature + '_encoded'] = label_encoder.fit_transform(df_enc[feature])
```

```python
df_enc.head()
```

```python
encoded_features = ["marital-status_encoded", "occupation_encoded", "relationship_encoded", "hours-per-week", "capital-gain", "capital-loss", "income_encoded"]
```

```python
clf5 = LogisticRegression(max_iter=200).fit(df_enc[encoded_features], df_enc["M/F"])
```

```python
acc = clf5.score(df_enc[encoded_features], df["M/F"])
print(f"Accuracy is {acc}.")
```

We now attempt to play around with the attributes, with the aim being to either improve the accuracy of the model, or not reduce its accuracy too much.

```python
def logistic_acc(df, features):
    clf = LogisticRegression(max_iter=200).fit(df[features], df["M/F"])
    acc = clf.score(df[features], df["M/F"])
    
    print(f"Accuracy of model is {acc}.")
    return clf
```

```python
result1 = logistic_acc(df_enc, encoded_features)
```

```python
logistic_acc(df_enc, ["marital-status_encoded", "occupation_encoded", "relationship_encoded", "hours-per-week", "income_encoded"])
```
